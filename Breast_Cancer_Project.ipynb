 ## Breast Cancer Dataset Classification Supervied Algorithms

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
bc=load_breast_cancer()
```

```python
bc.feature_names
```

```python
df_x= pd.DataFrame(bc.data, columns= bc.feature_names)
df_x.head(2)
```

```python
df_y = pd.DataFrame(bc.target, columns=['target'])
df_y.head()
```

```python
df = pd.concat([df_x, df_y], axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_train.head()
```

```python
df.head()
```

 ## Normalization

```python

from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler(feature_range=(0,1))
x_train= scalar.fit_transform(x_train)
x_train= pd.DataFrame(x_train)
x_test= scalar.transform(x_test)
x_test= pd.DataFrame(x_test)
x_test.head(2)
```

 ## 1- Naive Bayes

```python
# Classification Algorithms (Supervised)
# 1- Naive Bayes (Guassian Naive Bayes- GuassianNB)

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

 ## Create A Function For Evaluation Metrics
 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score 

def calculate_metrics(y_train,y_test, y_pred_train, y_pred_test):
    acc_train = accuracy_score(y_true=y_train, y_pred=y_pred_train)
    acc_test= accuracy_score(y_true=y_test, y_pred= y_pred_test)
    p= precision_score(y_true=y_test, y_pred= y_pred_test)
    r= recall_score(y_true=y_test, y_pred= y_pred_test)
    
    print(f"acc_train: {acc_train}, acc_test:{acc_test}, percision:{p}, recall:{r}")
    return acc_train, acc_test, p, r
```

```python
y_pred_train= gnb.predict(x_train)
y_pred_test= gnb.predict(x_test)

acc_train_gnb, acc_test_gnb, p_gnb, r_gnb = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 2- KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=8 , algorithm='kd_tree', leaf_size= 28)
knn.fit(x_train, y_train)
```

```python
y_pred_train= knn.predict(x_train)
y_pred_test= knn.predict(x_test)

acc_train_knn, acc_test_knn, p_knn, r_knn = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 3-Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=128, min_samples_split=4, criterion= 'entropy')

dt.fit(x_train, y_train)
```

```python
y_pred_train= dt.predict(x_train)
y_pred_test= dt.predict(x_test)

acc_train_dt, acc_test_dt, p_dt, r_dt = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 4-Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, max_depth=32, min_samples_split=4)
rf.fit(x_train, y_train)
```

```python
y_pred_train= rf.predict(x_train)
y_pred_test= rf.predict(x_test)

acc_train_rf, acc_test_rf, p_rf, r_rf = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 5- SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train, y_train)
```

```python
y_pred_train= svm.predict(x_train)
y_pred_test= svm.predict(x_test)

acc_train_svm, acc_test_svm, p_svm, r_svm= calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 6- Logistic Regression
 

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_train= lr.predict(x_train)
y_pred_test= lr.predict(x_test)

acc_train_lr, acc_test_lr, p_lr, r_lr = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## 7- ANN (Artificial Neural Network)

```python
from sklearn.neural_network import MLPClassifier
ann = MLPClassifier(hidden_layer_sizes=256, activation='relu', solver='adam', batch_size=64)
ann.fit(x_train, y_train)
```

```python
y_pred_train= ann.predict(x_train)
y_pred_test= ann.predict(x_test)

acc_train_ann, acc_test_ann, p_ann, r_ann = calculate_metrics(y_train,y_test, y_pred_train, y_pred_test)
```

 ## Comparison Algorithms

```python
import matplotlib.pyplot as plt
acc_train=[acc_train_gnb,acc_train_knn, acc_train_dt, acc_train_rf, acc_train_svm, acc_train_lr, acc_train_ann ]
algorithms_names= ['GNB','KNN', 'DT', 'RF', 'SVM', 'LR', 'ANN' ]
colors = ['Black','red', 'green', 'orange', 'purple', 'brown', 'blue']
plt.bar(algorithms_names, acc_train, color= colors)
plt.title('Evaluate Train Accuracy')
plt.grid()
```

```python
 # Test accuracy
acc_test=[acc_test_gnb,acc_test_knn, acc_test_dt, acc_test_rf, acc_test_svm, acc_test_lr, acc_test_ann ]
algorithms_names= ['GNB','KNN', 'DT', 'RF', 'SVM', 'LR', 'ANN' ]
colors = ['Black','red', 'green', 'orange', 'purple', 'brown', 'blue']
plt.bar(algorithms_names, acc_test, color= colors)
plt.title('Evaluate Test Accuracy')
plt.grid()
```

```python
# Percision
p_test=[p_gnb,p_knn, p_dt, p_rf, p_svm, p_lr, p_ann ]
algorithms_names= ['GNB','KNN', 'DT', 'RF', 'SVM', 'LR', 'ANN' ]
colors = ['Black','red', 'green', 'orange', 'purple', 'brown', 'blue']
plt.bar(algorithms_names, p_test, color= colors)
plt.title('Evaluate Test Percision')
plt.grid()
```

```python
# Recall
r_test=[r_gnb,r_knn, r_dt, r_rf, r_svm, r_lr, r_ann ]
algorithms_names= ['GNB','KNN', 'DT', 'RF', 'SVM', 'LR', 'ANN' ]
colors = ['Black','red', 'green', 'orange', 'purple', 'brown', 'blue']
plt.bar(algorithms_names, r_test, color= colors)
plt.title('Evaluate Test Recall')
plt.grid()
```
